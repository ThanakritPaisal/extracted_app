```python
import json
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager

url = "https://www.tiktok.com/api/post/item_list/?WebIdLastTime=1763439271&aid=1988&app_language=en&app_name=tiktok_web&browser_language=en-US&browser_name=Mozilla&browser_online=true&browser_platform=Linux%20x86_64&browser_version=5.0%20%28X11%3B%20Linux%20x86_64%29%20AppleWebKit%2F537.36%20%28KHTML%2C%20like%20Gecko%29%20HeadlessChrome%2F142.0.0.0%20Safari%2F537.36&channel=tiktok_web&cookie_enabled=true&count=35&coverFormat=2&cursor=0&data_collection_enabled=false&device_id=7573913970912151058&device_platform=web_pc&focus_state=true&history_len=2&is_fullscreen=false&is_page_visible=true&language=en&odinId=7573913507822961681&os=linux&priority_region=&referer=&region=TH&screen_height=600&screen_width=800&secUid=MS4wLjABAAAA8i0IUFsQtz7WH4UT_F536IRUPrxnyf6uommbBZbkOt48MOLIadJ6sJ0CTf87y9qm&tz_name=UTC&user_is_login=false&video_encoding=mp4&webcast_language=en&msToken=&X-Bogus=DFSzsIVOe9zANVY-COS9JCvDKKMP&X-Gnarly=MxZ137pK70ealNbYlx2yxZsAVuuwffkwX8bUK9X3fOTggB6cO1aoDjcMvQZapGnpjRa7UGLUOvJwb6aV7npKPYZ2obXeP7Tvo8yjrsH998ZjYSqwbk7wsGFQ6VhtMRxhoPRL1elfSYWQlUbZvcZtRvHB6KKnW5-/bK5eDaFbtoaPLKDhmwqs3CzxKmHd1f8utqBP0eyYndqKFGzMeZjZY6KX1-VtWfGdGZw80uSbX3yzUbt4jSSqvc-BfpS8Bf6mNZfQ19MksX7vH5guOjIIamWkMwVStr7vD43yERDrN3u6OjMxUoDbiTLJUq89zeQVwqZ="

# ตั้งค่าบราวเซอร์ (ใช้ headless ได้)
options = Options()
options.add_argument("--headless=new")
options.add_argument("--disable-gpu")
options.add_argument("--no-sandbox")

driver = webdriver.Chrome(
    service=Service(ChromeDriverManager().install()),
    options=options,
)

try:
    driver.get(url)

    # ถ้า endpoint นี้คืน JSON ตรง ๆ ส่วนมาก Chrome จะ render เป็น <pre> tag
    raw_text = ""
    try:
        pre = driver.find_element(By.TAG_NAME, "pre")
        raw_text = pre.text
    except Exception:
        # fallback ถ้าไม่มี <pre> ก็ใช้ page_source แทน
        raw_text = driver.page_source

    print("==== RAW TEXT ====")
    print(raw_text[:500])  # preview 500 ตัวอักษรแรก

    # ถ้าคิดว่าเป็น JSON จริง ๆ ลอง parse ต่อได้
    try:
        data = json.loads(raw_text)
        print("Parsed JSON keys:", list(data.keys())[:20])
    except Exception as e:
        print("ไม่สามารถ parse JSON:", e)

finally:
    driver.quit()
```
